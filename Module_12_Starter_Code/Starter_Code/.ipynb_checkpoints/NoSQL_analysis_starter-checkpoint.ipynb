### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene' : 20}
# Use count_documents to display the number of documents in the result
result = establishments.find(query)

hygiene_result = []
for i in result:
    hygiene_result.append(i)
# Display the first document in the results using pprint
```

```python
# Convert the result to a Pandas DataFrame
df_hygiene = pd.json_normalize(hygiene_result)
```

```python
# Display the first 10 rows of the DataFrame
df_hygiene.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'LocalAuthorityName' : {'$regex' : 'London'}, 'RatingValue' : {'$gte' : '4'}}
results = establishments.find(query)

# Use count_documents to display the number of documents in the result
print(f"there are {establishments.count_documents(query)} establishments in London with a rating value greater or equal to 4")
# Display the first document in the results using pprint

pprint(establishments.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
df = pd.json_normalize(results)
```

```python
# Display the number of rows in the DataFramed
print(f"there are {len(df['_id'])} rows in the dataframe")
```

```python
# Display the first 10 rows of the DataFrame
df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
query = {'BusinessName' : 'Penang Flavours'}
pprint(establishments.find_one(query))
print(establishments.find_one(query)['geocode']['latitude'])
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
import numpy as np

degree_search = 0.01
latitude = establishments.find_one(query)['geocode']['latitude']
longitude = establishments.find_one(query)['geocode']['longitude']
```

```python
lon_range = [longitude-degree_search, longitude+degree_search]
lat_range = [latitude-degree_search, latitude+degree_search]
```

```python
# search must only include latitude+degree_search, latitude-degree_search, longitude+degree_search, and longitude-degree_search
nearby = {'scores.Hygiene': {'$eq' :5}, 'geocode.longitude' : {'$gte' : lon_range[0], '$lte' : lon_range[1]},
          'geocode.latitude' : {'$gte' : lat_range[0], '$lte' : lat_range[1]}}
sort =  [('scores.Hygiene' ,1)]

# Print the results
nearby_results = establishments.find(nearby).sort(sort)
```

```python
nearby_res = []
for i in nearby_results:
    nearby_res.append(i)
```

```python
# Convert result to Pandas DataFrame
nearby_df = pd.json_normalize(nearby_res)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
hygiene_0_query = {'$match' : {'scores.Hygiene' : {'$eq' : 0}}}
#hygiene_0_query = {'scores.Hygiene' : 0}

# 2. Groups the matches by Local Authority
group_query = {'$group' : {'_id' : '$LocalAuthorityName', 'count' : {'$sum' : 1}}}

# 3. Sorts the matches from highest to lowest
sort_values = {'$sort' : {'count': -1}}
```

```python
#putting the pipeline together
pipeline = [hygiene_0_query, group_query, sort_values]
```

```python
#creating a result variable
results = list(establishments.aggregate(pipeline))
```

```python
# Print the number of documents in the result
print(f'there are {len(results)} documents in the result')

# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
grouped_df = pd.json_normalize(results)
# Display the number of rows in the DataFrame
print(f"there are {len(grouped_df.index)} rows in this dataframe")
# Display the first 10 rows of the DataFrame
grouped_df.head(10)
```
